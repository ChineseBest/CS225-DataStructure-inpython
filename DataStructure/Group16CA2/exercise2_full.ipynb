```python
#Exercise 2 part
#This query part tests two queries, which are related to (i) and (iii).
import copy
import time
class Relation:
    class smallBlock:
        def __init__(self,size):
            self.size = size
            self.tuples = [None] * size
            self.numtuples = 0
        def __getitem__(self,idx):
            return self.tuples[idx]
        def __setitem__(self,idx,t):
            self.tuples[idx] = t
        def getnumtuples(self):
            return self.numtuples
        def insert(self,t):
            for i in range(self.size):
                if self.tuples[i] == None:
                    self.tuples[i] = t
                    self.numtuples += 1
                    return i
            raise IndexError("The block is full.")

        def retrieve(self,idx):
            return self.tuples[idx]
        
        def isfull(self):
            count=0
            for i in range(self.size):
                if self.tuples[i] != None:
                    count +=1
            if count==self.size-1:
                return True
            return False

    class __Placeholder:
        def __init__(self):
            pass
        def __eq__(self,other):
            return False

    def __init__(self,thesize,key):
        self.size=thesize
        self.elements=[None]*self.size
        self.key=key
        #idlist, to help travel.
        self.idxlist=[]
        for i in range(0,self.size//2):
            self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]  

    def insert(self,t):
        if t[0] in self.idxlist:
            raise KeyError('data already exists')
        self.idxlist.append(t[0])
        idx1=t[0]%40
        idx2=t[0]//40
        if self.elements[idx2][0].isfull()== False:
            self.elements[idx2][0][idx1]=t
            self.elements[idx2][0].numtuples+=1

        if self.elements[idx2][0].isfull()== True:
            self.sorting(idx2)
            self.split(self.elements[idx2][1],idx2)
        return self.findpointer(t)
        
    
    def deletion(self,t):
        flag=0
        for i in self.elements:
            if i != None:
                for j in i[0].tuples:
                    if t==j:
                        flag=1
                        num=i[0].tuples.index(j)
                        i[0].tuples[num]=Relation.__Placeholder()
                        i[0].numtuples-=1
                        self.idxlist.remove(j[0])
                        if i[0].numtuples ==i[0].size//2:
                            self.merge(i)
                        return True

                for k in i[1].tuples:
                    if t==k:
                        flag=1
                        num=i[1].tuples.index(k)
                        self.idxlist.remove(k[0])
                        i[1].tuples[num]=Relation.__Placeholder()
                        i[1].numtuples-=1
                        if i[1].numtuples ==i[1].size//2:
                            self.merge(i)
                        return True
        if flag==1:
            return True
            
        if flag==0:
            raise IndexError('No such key')
    
    def keydelete(self,key):
        element=self.find(key)
        self.deletion(element)


    
    def sorting(self,idx2):
        list0=[]
        for i in self.elements[idx2][0].tuples:
            if i!=None and type(i)!=Relation.__Placeholder:
                list0.append(i)
        for i in self.elements[idx2][1].tuples:
            if i!=None and type(i)!=Relation.__Placeholder:
                list0.append(i)

        #Lots of differnet primary keys
        if self.key=='Movie':
            list0.sort(key=lambda x:(x[1],x[2]))

        if self.key=='Person':
            list0.sort(key=lambda x:x[1])
        
        if self.key=='Award':
            list0.sort(key=lambda x:x[1])
        
        if self.key=='Restriction_Category':
            list0.sort(key=lambda x:(x[1],x[2]))
        
        if self.key=='Director':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Writer':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Crew':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Role':
            list0.sort(key=lambda x:(x[2],x[3],x[4]))
        
        if self.key=='Scene':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Restriction':
            list0.sort(key=lambda x:(x[1],x[2],x[3],x[4]))
        
        if self.key=='Appearance':
            list0.sort(key=lambda x:(x[1],x[2],x[3],x[4]))
        
        if self.key=='Movie_Award':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Crew_Award':
            list0.sort(key=lambda x:(x[1],x[2],x[3],x[4]))
        
        if self.key=='Director_Award':
            list0.sort(key=lambda x:(x[1],x[2],x[3]))
        
        if self.key=='Writer_Award':
            list0.sort(key=lambda x:(x[1],x[2],x[3],x[4]))
        
        if self.key=='Actor_Award':
            list0.sort(key=lambda x:(x[1],x[2],x[3],x[4]))
            
        if self.key=='QMovie_Award':
            list0.sort(key=lambda x:(list(x[2].values())[0][2],list(x[2].values())[1][2],list(x[2].values())[2][2]))

        if self.key=='QPerson':
            list0.sort(key=lambda x:list(x[2].values())[0][2])
        
        if self.key=='QRole':
            list0.sort(key=lambda x:(list(x[2].values())[1][2],list(x[2].values())[2][2],list(x[2].values())[3][2]))
        
        if self.key=='QScene':
            list0.sort(key=lambda x:(list(x[2].values())[0][2],list(x[2].values())[1][2],list(x[2].values())[2][2]))
        
        if self.key=='QAppearance':
            list0.sort(key=lambda x:(list(x[2].values())[0][2],list(x[2].values())[1][2],list(x[2].values())[2][2],list(x[2].values())[3][2]))

        if self.key=='QRestriction':
            list0.sort(key=lambda x:(list(x[2].values())[0][2],list(x[2].values())[1][2]))


        self.elements[idx2][0]=Relation.smallBlock(40)
        self.elements[idx2][1]=Relation.smallBlock(40)
        for j in list0:
            self.elements[idx2][1].insert(j)
        
    
    def split(self,elements,idx):
        for i in range(self.size//2,self.size):
            if self.elements[i]==None:
                self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]
                for j in range(self.elements[i][0].size//2,self.elements[i][0].size):
                    self.elements[i][0].insert(elements[j])
                    self.elements[idx][1][j]=None
                    self.elements[idx][1].numtuples-=1
                return i 
                    
    def merge(self,tuplea):
        list1=[]
        for i in tuplea[0].tuples:
            if i!= None:
                list1.append(i)
        for i in tuplea[1].tuples:
            if i!= None:
                list1.append(i)
        for i in range(0,self.size):
            if self.elements[i]==tuplea:
                if i==0:
                    for k in list1:
                        self.elements[1][0].insert(k)
                    self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]
                    return
                if i==self.size-1:
                    for k in list1:
                        self.elements[self.size-1][0].insert(k)
                    self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]
                    return
                if self.elements[i-1][1].numtuples>=self.elements[i+1][1].numtuples:
                    for k in list1:
                        self.elements[i+1][0].insert(k)
                    self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]
                    return
                if self.elements[i-1][1].numtuples < self.elements[i+1][1].numtuples:
                    for k in list1:
                        self.elements[i-1][0].insert(k)
                    self.elements[i]=[Relation.smallBlock(40),Relation.smallBlock(40)]
                    return

    #This function is same as find/ another kind of retireval.
    def find(self,num):
        #Linear search
        for i in self.elements:
            if i != None:
                for j in i[0].tuples:
                    if j!=None and type(j)!=Relation.__Placeholder and num == j[0]:
                        return j
                couplelist=[]
                for j in i[1].tuples:
                    if j!= None and type(j)!=Relation.__Placeholder:
                        couplelist.append(j)
                if couplelist ==[]:
                    continue
                couplelist.sort(key=lambda x:x[0])
                left=0
                right=len(couplelist)
                #binary search
                while left<=right:
                    mid=(left+right)//2
                    if num<couplelist[mid-1][0]:
                        right=mid-1
                    elif num>couplelist[mid-1][0]:
                        left=mid+1
                    else:
                        return couplelist[mid-1]
        return -1
    
    def retrievedirectly(self,pointer):
        blocknumber = pointer[0]
        idx = pointer[1]
        blocktype=pointer[2]
        return self.elements[blocknumber][blocktype][idx]
    
    def findpointer(self,t):
        for i in range(len(self.elements)):
            if self.elements[i] != None:
                for j in range(self.elements[i][0].size):
                    if self.elements[i][0][j]==t:
                        return (i,j,0)
                for j in range(self.elements[i][1].size):
                    if self.elements[i][1][j]==t:
                        return (i,j,1)
```

```python
#Class Stack, implemented in CA1.
class Stack:
    def __init__(self,size=20):
        self.items = [None] * size
        self.numItems = 0
        self.size = size
        
    def top(self):
        if self.numItems != 0:
            return self.items[self.numItems-1]
        raise IndexError("Stack is empty")
        
    def push(self,item):
        if self.numItems >= self.size:
            self.allocate()
        self.items[self.numItems] = item
        self.numItems += 1
        
    def allocate(self):
        newlength = 2 * self.size
        newStack = [None] * newlength
        for i in range(self.numItems):
            newStack[i] = self.items[i]
        self.items = newStack
        self.size = newlength
        
    def pop(self):
        if self.numItems != 0:
            topelement = self.items[self.numItems-1]
            self.numItems -= 1
            return topelement
        raise IndexError("Stack is empty")
        
    def deallocate(self):
        newlength = self.size //2
        newStack = [None] * newlength
        for i in range(self.numItems):
            newStack[i] = self.items[i]
        self.items = newStack
        self.size = newlength
        
    def isEmpty(self):
        if self.numItems != 0:
            return False
        return True

    def search(self,n):
        for i in range(self.numItems):
            if n == self.items[i]:
                return True
        return False

#Create instance。
RES=Stack(100)
ENV=Stack(100)

#Crate key list for our database.
namelist=[]
for i in range(1,21):
    namelist.append(i)


#Functionlist
def dereference(ident):
    for i in database.keys():
        if i != ident:
            for j in database[i][2].keys():
                if ident == j:
                    return database[i][2][j][2]
        else:
            return list(database[i][2].keys())

def findname(ident):
    for i in database.keys():
        if i != ident:
            for j in database[i][2].keys():
                if ident == j:
                    return database[i][2][j][1]
        else:
            return database[i][1]

def nametestbigtitle(name):
    bigtitle=[]
    for i in database.keys():
        if database[i][1] == name[0]:
            bigtitle.append(i)
    if bigtitle != []:
        RES.push(bigtitle)
        return True
    return False

def nametestsmalltitle(name):
    smalltitle=[]
    for i in database.keys():
        for j in database[i][2].keys():
            if database[i][2][j][1] == name[0]:
                smalltitle.append(j)
    if smalltitle != []:
        ENV.push(smalltitle)
        return True
    return False

def valuetest(name):
    valueresult=[]
    for i in database.keys():
        if dereference(i) == name[0]:
            valueresult.append(i)
        else:
            for j in database[i][2].keys():
                if dereference(j) == name[0]:
                    valueresult.append(j)
    if valueresult != []:
        ENV.push(valueresult)
        return True    
    return False


def distinct(result):
    return set(result)
    

def In(a,b):
    for genre in b:
        for name in a:
            if name in database[genre][2].keys():
                return True

    return False
    
def Contains(a,b):
    seta = set(a)
    setb = set(b)
    return seta.issubset(setb)

def Exist(a):
    if a == []:
        return False
    return True

#Functions mainly work in query q1
def testmoviename(query):
    if 'Movie' in query[0]:
        return True
    return False

def testsceneame(query):
    if 'scene' in query[0]:
        return True
    return False

#Distinguish it from the dereference () function.
def dereferencenew(ident,database):
    for i in database.keys():
        if i != ident:
            for j in database[i][2].keys():
                if ident == j:
                    return database[i][2][j][2]
        else:
            return list(database[i][2].keys())

#Find actors that occur in same scene.
def findcommonapp():
    database=eval('Appearance')
    list0=[]
    for i in database.keys():
        dict1=database[i][2]
        listkey=list(dict1.keys())
        id=dict1[listkey[3]][2]
        name=dict1[listkey[2]][2]
        list0.append((id,name))  
    return list0

#Find the number of scenes that the specific actor occurs.
def findapp(actor,list):
    result=[]
    for i in list:
        if i[1]==actor:
            result.append(i[0])
    return result

#Find the actors that donot play in the same scene.
def findcommonact():
    appearancelist=findcommonapp()
    actor=set()
    scenelist={}
    for i in appearancelist:
        actor.add(i[1])
    actor=list(actor)
    actor.sort()
    for i in actor:
        scenelist[i]=set(findapp(i,appearancelist))
    result=[]
    for i in actor:
        actors=copy.deepcopy(actor)
        actors.remove(i)
        for j in actors:
            if scenelist[i].intersection(scenelist[j])==set():
                result.append((i,j))
    return result

    
#Use this to check the id for actors.
#Usually it is used to help debug.
def findcommonperson():
    result=findcommonact()
    ENV.push(namelist)
    keyname=[]
    listkey=[]
    database=Role
    for j in Role.keys():
        listkey.append(dereferencenew(j,database))
    for i in result:
        main=i[0]
        amior=i[1]
        for k in listkey:
            if main==dereferencenew(k[3],database):
                main=dereferencenew(k[0],database)
            if amior==dereferencenew(k[3],database):
                amior=dereferencenew(k[0],database)
        keyname.append((main,amior))
    for key in keyname:
        if (key[1],key[0]) in keyname:
            keyname.remove((key[1],key[0]))
    RES.push(keyname)

#The last step, to change actors into real name.
def gotfinalactor():
    keyname=RES.pop()
    database=Person
    personlist=[]
    finalresult=[]
    for i in database.keys():
        personlist.append(dereferencenew(i,database))
    for i in keyname:
        main=i[0]
        amior=i[1]
        for k in personlist:
            if main==dereferencenew(k[0],database):
                main=dereferencenew(k[1],database)+' '+dereferencenew(k[2],database)
            if amior==dereferencenew(k[0],database):
                amior=dereferencenew(k[1],database)+' '+dereferencenew(k[2],database)
        finalresult.append((main,amior))
    RES.push(finalresult)

#Functions mainly work in query q2
def testrestriction(query):
    if 'Restriction' in query[0]:
        return True
    return False

def testrole(query):
    if 'role' in query[0]:
        return True
    return False

#Find the movie that not suitable for children.
def pickcrime(database):
    ENV.push(namelist)
    crimemovie=[]
    for i in database.keys():
        crimelist=dereferencenew(i,database)
        crimemovie.append(dereferencenew(crimelist[0],database))
    RES.push(crimemovie)

#Choose the actors that act in movies suitable for children.
def choosesutiable(database):
    crimemovie=RES.pop()
    actorlist=[]
    for i in database.keys():
        crimelist=dereferencenew(i,database)
        if dereferencenew(crimelist[1],database) in crimemovie:
            actorlist.append(dereferencenew(crimelist[3],database))
    actorlist=list(set(actorlist))
    actorlist.sort()
    RES.push(actorlist)

#eval() function, from CA1
def evalnew(query):
    if len(query) == 3:
        evalnew(query[0])
        evalnew(query[2])
        if query[1] == 'where':
            comp1=RES.pop()
            comp2=RES.pop()
            result=[]
            for i in comp2:
                for j in comp1:
                    if j in dereference(i):
                        result.append(i)
            RES.push(result)
            print('selection result',result)
            return

        if query[1] == 'satisfy':
            res1=RES.pop()
            ENV.pop()
            print('satisfy result')
            print(res1)


        if query[1] == '=':
            comp1=distinct(ENV.pop())
            comp2=distinct(ENV.pop())
            comp3=list(comp1.intersection(comp2))
            RES.push(comp3)
            return

        if query[1] == '.':
            result=[]
            comp1=ENV.pop()
            print(comp1)
            comp2=RES.pop()
            print(comp2)
            for i in comp1:
                for j in comp2:
                    if i in dereference(j):
                        result.append(i)
            RES.push(result)
            print('Q.A attribution',result)
            return

        if query[1] == 'eq':
            comp0=[]
            comp1=RES.pop()
            comp2=RES.pop()
            for i in comp1:
                for j in comp2:
                    if dereference(i) == dereference(j):
                        comp0.append([i,j])
            RES.push(comp0)
            print('eq result',comp0)
            return

        if query[1] == 'EQJ':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            for i in comp1:
                for j in comp2:
                    i=set(i)
                    j=set(j)
                    if j.intersection(i)== i:
                        res.append(list(j))
            RES.push(res)
            print('EQJOIN',res)
            return
          
        if query[1] == 'x':
            res1=[]
            res2=[]
            res =[]
            comp1=RES.pop()
            comp2=RES.pop()
            for i in comp1:
                res1.append(dereference(i))
            for j in comp2:
                res2.append(dereference(j))
            for a in res1:
                for b in res2:
                    res.append(a+b)
            RES.push(res)
            print('Cartesian',res)
            return

            
        if query[1] == '+':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]+comp1[0])
            RES.push(res)
            print('+',res)
            return

        if query[1] == '-':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]-comp1[0])
            RES.push(res)
            print('-',res)
            return

        if query[1] == '/':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]/comp1[0])
            RES.push(res)
            print('/',res)
            return

        if query[1] == '*':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]*comp1[0])
            RES.push(res)
            print('*',res)   
        
        if query[1] == '>':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]>comp1[0])
            RES.push(res)
            print('>',res) 

        if query[1] == '<':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]<comp1[0])
            RES.push(res)
            print('<',res)  
        
        if query[1] == '==':
            res=[]
            comp1=RES.pop()
            comp2=RES.pop()
            res.append(comp2[0]==comp1[0])
            RES.push(res)
            print('==',res)  

        if query[1] == 'in':
            b = RES.pop()
            a = ENV.pop()
            res = In(a,b)
            print('in This statement is:',res)
            RES.push(res)

        if query[1] == 'contained in':
            b = ENV.pop()
            a = ENV.pop()
            res = Contains(a,b)
            print('contains This statement is:',res)
            RES.push(res)         

    
    if len(query) == 2:
        evalnew(query[1])#For next level, recursively call eval
        if query[0] == 'Count':#For the operator is 'Count()'
            comp1=RES.pop()
            length = [len(comp1)]
            RES.push(length)
            print(length)
        if query[0] == 'Sum':#For the operator is 'Sum'
            comp1=RES.pop()
            comp1=dereference(dereference(comp1[0])[0])
            for i in comp1:
                a += i
            RES.push([a])
            print('Sum',a)
        if query[0] == 'Average':#For the operator is 'Average'
            comp1=RES.pop()
            comp1=dereference(dereference(comp1[0])[0])
            for i in comp1:
                a += i
            a = a / len(comp1)
            RES.push([a])
            print('Average',a)
        if query[0] == 'Min':#For the operator is 'Min'
            comp1=RES.pop()
            comp1=dereference(dereference(comp1[0])[0])
            a = comp1[0]
            for i in comp1:
                if i < a:
                    a = i
            RES.push([a])  
            print('Min',a)              
        if query[0] == 'Max':#For the operator is 'Max'
            comp1=RES.pop()
            comp1=dereference(dereference(comp1[0])[0])
            a = comp1[0]
            for i in comp1:
                if i > a:
                    a = i
            RES.push([a])
            print('Max',a)

        if query[0] == 'exist':
            a = RES.pop()
            res = Exist(a)
            print("exist This statement is:",res)
            RES.push(res)
            

    if len(query) == 1:

        if testmoviename(query):
            findcommonperson()
            return
        if testsceneame(query):
            gotfinalactor()
            return 

        if testrestriction(query):
            pickcrime(Restriction)
            return

        if testrole(query):
            choosesutiable(Role)
            return

        if nametestbigtitle(query):
            return
        
        if nametestsmalltitle(query):
            return 

        if valuetest(query):
            return
```
